```python
import pandas as pd
df = pd.read_csv("astronauts.csv")
df.head()
```

# modifying column data type

```python
birthdate = pd.to_datetime(df["Birth Date"], format = "%m/%d/%Y") # format = "%m/%d/%Y" is optional, 
#only need to specify if not detected automatically
birthdate
```

```python
# its useful to convert to datetime because it allows for easy manipulattion e.g. to get year:
birthdate.dt.year
# to get day - birthdate.dt.day
# to get month - birthdate.dt.month
```

```python
# converting to datetime allows for easy calculations with dates:
zarya = pd.to_datetime("1998-11-20")
df["age_at_zarya"] = (zarya-birthdate).astype("timedelta64[Y]")
df.head(3)
```

```python
df["birth"] = birthdate
df.head(3)
```

# Apply, Map & Vectorized functions

```python
data = np.round(np.random.normal(size=(4, 3)), 2)
df = pd.DataFrame(data, columns = ["A", "B", "C"])
df.head()
```

```python
# add 1 to every value in dataframe with .apply()
df.apply(lambda x: 1 + np.abs(x))
```

```python
# .map() it is similar to apply, but operates on Series ONLY and uses dictionary based input rather than 
# an array of values
```

```python
series = pd.Series(["Steve", "Alex", "Jess", "Mark"])
```

```python
series.map({"Steve": "Steven"})
```

```python
# or with a function
series.map(lambda d: f"I am {d}")
```
