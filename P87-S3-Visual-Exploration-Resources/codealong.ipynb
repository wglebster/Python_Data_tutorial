#Basics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv("heart.csv")
```

#barplots

```python
df.head()
```

```python
chest_pain = df.groupby(by="cp").median().reset_index()
chest_pain.head(10)
```

```python
chest_pain.plot.bar(x="cp", y="age"); #semicolon at the end suppresses additional
#output above the plot
```

```python
chest_pain.plot.bar(x="cp");
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain["cp"], chest_pain["age"], label = "age")
ax.set_xlabel("cp")
ax.legend();
```

```python
ax = sb.barplot(x="cp", y="age", capsize = 0.1, data = df)
```

#scatterplots

```python
df.plot.scatter(x="age", y="trestbps");
```

```python
fig.ax = plt.subplots()
ax.scatter(df["age"], df["trestbps"], marker="^", 
           s=60, c=df["age"], edgecolors = "k")
ax.set_xlabel("age")
ax.set_ylabel("trestbps");
```

```python
sb.scatterplot(x="age", y="trestbps", hue = "age", data = df);
```

#lineplots

```python
ages = df.groupby("age").median().reset_index()
ages.head()
```

```python
ages.plot.line("age", "chol");
```

```python
ages.plot.line("age", ["chol", "trestbps"]);
```

```python
fig, ax = plt.subplots()
ax.plot(ages["age"], ages["trestbps"], ls = ":", lw = 2)
ax.set_xlabel("age")
ax.set_ylabel("trestbps");
```

```python
fig, axes = plt.subplots(ncols=2)
df.plot.scatter(x="age", y="chol", ax=axes[0])
df.plot.scatter(x="age", y="trestbps", ax = axes[1])
fig.tight_layout();
```

#distributions

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv("heart.csv")
df.head()
```

#histograms

```python
df.age.plot.hist(bins=50);
```

```python
plt.hist(df.age, bins=50);
```

```python
fix, axis = plt.subplots()
axis.hist(df.trestbps, bins = 20, histtype = "step", label = "trestbps")
axis.hist(df.thalach, bins = 20, histtype = "stepfilled", label = "thalach")
axis.legend();
```

#boxplots

```python
df[["trestbps", "thalach"]].plot.box();
df[["trestbps", "thalach"]].describe()
```

```python
plt.boxplot(df[["trestbps", "thalach"]].to_numpy());
```

```python
sb.boxplot(data = df[["trestbps", "thalach"]]);
```

```python
sb.boxplot(data = df, x="cp", y="trestbps");
```

violin plots

```python
fig, axis = plt.subplots()
axis.violinplot(df[["trestbps", "thalach"]].to_numpy());
```

```python
fig, axis = plt.subplots()
axis.violinplot(df[["trestbps", "thalach"]].to_numpy(), bw_method=0.2);
```

```python
sb.violinplot(data = df[["trestbps", "thalach"]], inner = "quartile", bw = 0.2);
```

#swarmplot

```python
sb.swarmplot(data = df[["trestbps", "thalach"]], size=2, color="black", alpha = 0.5);
sb.violinplot(data = df[["trestbps", "thalach"]]);
```

#2d vizualisations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

df = pd.read_csv("meteorite-landings.csv")
df.head()
```

```python
df.info()
```

```python
df.dropna(subset = ["reclat", "reclong"])
df = df[df.reclong<300]
```

```python
df.info()
```

```python
plt.hist2d(x=df["reclong"], 
           y=df["reclat"], bins = 200, vmax = 10)
plt.colorbar();
```

```python
plt.hexbin(x=df["reclong"], 
           y=df["reclat"], bins = 200, vmax = 10, lw=0.0)
```

```python
df.plot.hexbin(x="reclong", y="reclat", vmax=2, gridsize=100, linewidth=0.25)
```
